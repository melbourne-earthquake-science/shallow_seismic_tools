```python
from obspy.imaging.cm import obspy_sequential
from scipy.signal import spectrogram
def _nearest_pow_2(x):
    """Returns the nearest power of 2 greater than or equal to x."""
    return 2 ** int(np.ceil(np.log2(x)))


def spectrogram2(data, samp_rate, per_lap=0.9, wlen=None, log=False,
                       outfile=None, fmt=None, axes=None, dbscale=False,
                       mult=8.0, cmap=plt.cm.viridis, zorder=None, title=None,
                       show=True, clip=[0.0, 1.0], freqmin=None, freqmax=None):
    """
    Computes and plots spectrogram of the input data.
    
    :param freqmin: Minimum frequency for the plot
    :param freqmax: Maximum frequency for the plot
    """
    # enforce float for samp_rate
    samp_rate = float(samp_rate)

    # set wlen from samp_rate if not specified otherwise
    if not wlen:
        wlen = 128 / samp_rate

    npts = len(data)
    nfft = int(_nearest_pow_2(wlen * samp_rate))

    if npts < nfft:
        msg = (f'Input signal too short ({npts} samples, window length '
               f'{wlen} seconds, nfft {nfft} samples, sampling rate '
               f'{samp_rate} Hz)')
        raise ValueError(msg)

    nlap = int(nfft * float(per_lap))

    data = data - data.mean()
    end = npts / samp_rate

    # Use scipy.signal.spectrogram to compute the spectrogram
    freq, time, specgram = spectrogram(data, fs=samp_rate, nperseg=nfft, noverlap=nlap)

    if dbscale:
        specgram = 10 * np.log10(specgram)
    else:
        specgram = np.sqrt(specgram)

    # If frequency limits are provided, filter the frequency range
    if freqmin or freqmax:
        freq_mask = np.ones_like(freq, dtype=bool)
        if freqmin:
            freq_mask &= freq >= freqmin
        if freqmax:
            freq_mask &= freq <= freqmax

        freq = freq[freq_mask]
        specgram = specgram[freq_mask, :]

    vmin, vmax = clip
    if vmin < 0 or vmax > 1 or vmin >= vmax:
        msg = "Invalid parameters for clip option."
        raise ValueError(msg)
    _range = float(specgram.max() - specgram.min())
    vmin = specgram.min() + vmin * _range
    vmax = specgram.min() + vmax * _range
    norm = plt.Normalize(vmin, vmax, clip=True)

    if not axes:
        fig = plt.figure()
        ax = fig.add_subplot(111)
    else:
        ax = axes

    halfbin_time = (time[1] - time[0]) / 2.0
    halfbin_freq = (freq[1] - freq[0]) / 2.0

    kwargs = {'cmap': cmap, 'zorder': zorder}
    if log:
        ax.set_yscale('log')
    ax.pcolormesh(time, freq, specgram, norm=norm, **kwargs)

    ax.axis('tight')
    ax.set_xlim(0, end)
    ax.grid(False)

    if axes:
        return ax

    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Frequency [Hz]')
    if title:
        ax.set_title(title)

    if outfile:
        if fmt:
            fig.savefig(outfile, format=fmt)
        else:
            fig.savefig(outfile)
    elif show:
        plt.show()
    else:
        return fig
```

```python
# Create subplots: 1 row, 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


spectrogram2(tr0.data, samp_rate = 1000, log=False, show=True, dbscale=True, axes = ax1, freqmin=0.0, freqmax=50, 
             mult=1.0, cmap=plt.cm.coolwarm)


spectrogram2(tr2.data, samp_rate = 1000, log=False, show=True, dbscale=True, axes = ax2, freqmin=0.0, freqmax=50, 
             mult=1.0, cmap=plt.cm.coolwarm)

# Adjust layout
plt.tight_layout()
plt.show()


```

```python
# Assuming `tr2` is your trace and has data from the geophone record
# Sampling rate from the trace
fs = tr2.stats.sampling_rate  # 1000 Hz in your case

# Compute the Power Spectral Density (PSD) using Welch's method
# nperseg controls the length of each segment; adjust as needed

freqs0, psd0 = welch(tr0.data, fs=fs, nperseg=1024)
freqs1, psd1 = welch(tr1.data, fs=fs, nperseg=1024)

freqs2, psd2 = welch(tr2.data, fs=fs, nperseg=1024)

# Plot the PSD
plt.figure(figsize=(10, 6))
plt.semilogy(freqs0, psd0)
plt.semilogy(freqs1, psd1)
plt.semilogy(freqs2, psd2, ls = '--')
plt.xlim(0.1, 10)  # Focus on the 1-100 Hz range
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.title('Power Spectral Density (Welch) of Active Source Geophone Record')
plt.grid(True, which='both', ls='--')
plt.show()
```
